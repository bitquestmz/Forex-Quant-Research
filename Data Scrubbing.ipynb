This specific cell below scrapes from the excel file and compiles all the data into a master dataframe. Depending on how many years of data you're uploading, it could take a bit.

```python
excel_files = os.listdir()            # lists directory w/ data as the working directory

# sets the column names for the master dataframe
column_names = ['Date',               # 
                'Open',           # 
                'High',           # 
                'Low',            # 
                'Close',          
                'Volume']

# create empty dataframe to append data to
master_df = pd.DataFrame(columns = column_names)  

for file in excel_files:
    curr_df = pd.read_excel(file)                    # read excel file into a temporary dataframe
    curr_df.columns = column_names                   # insert column names into dataframe (needed in order to append)
    master_df = master_df.append(curr_df,            # append temporary dataframe onto master dataframe
                                 ignore_index=True,
                                 sort=False)
```

## Select and Display Data

```python
# Start Period
start_year = '2019'
start_month = '12'
start_day = '19'
start_time = '12:00:00'
start = start_year + '-' + start_month + '-' + start_day + ' ' + start_time

# End Period
end_year = '2019'
end_month = '12'
end_day = '20'
end_time = '12:00:00'
end = end_year + '-' + end_month + '-' + end_day + ' ' + end_time

select_df = master_df
select_df = select_df.set_index('Date')
select_df = select_df[start:end]
```
